```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
orig_image = cv.imread("image0.jpg", 0)

orig_image.shape
```

```python
mask = np.zeros(orig_image.shape, dtype=np.uint8)
orig_image.shape
```

```python
mask[300:600, 180:580] = 255
```

```python
orig_image.shape
```

```python
cv.imwrite("image00.jpg", mask)
```

```python
imaged_masked = cv.bitwise_and(orig_image, orig_image, mask=mask)
```

```python
newImage = cv.inpaint(orig_image, mask, 7, flags=cv.INPAINT_TELEA)
```

```python
cv.imwrite("image2.jpg", newImage)
```

```python
plt.imshow(cv.cvtColor(newImage, cv.COLOR_BGR2RGB))
plt.show()
```
