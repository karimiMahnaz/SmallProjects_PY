```python
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
orig_image = cv.imread("image1.jpg")
orig_image_gray = cv.imread("image1.jpg",0)
```

```python
_ , mask0 = cv.threshold(orig_image_gray, 420, 755, cv.THRESH_BINARY)
```

```python
newImage = cv.inpaint(orig_image, mask0, 7, flags=cv.INPAINT_TELEA)
```

```python
plt.imshow(cv.cvtColor(newImage, cv.COLOR_BGR2RGB))
plt.show()
```
