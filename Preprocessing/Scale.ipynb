```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_raw  = pd.read_csv("heart.csv")
data = data_raw.copy()
```

```python
data.head()
```

```python
data.Age.hist(bins= 20)
plt.show()
```

# Max Normalize

```python
max_age = data.Age / data.Age.max()
```

```python
max_age.std()
```

```python
# between 0 and 1

max_age.hist()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
mm_age = scaler.fit_transform(data[["Age"]])
```

```python
plt.hist(mm_age)
```

```python
mm_age.min()
```

```python
mm_age.max()
```

```python
mm_age.std()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
s_scaler = StandardScaler()
standard_age = s_scaler.fit_transform(data[["Age"]])
```

```python
standard_age.min()
```

```python
standard_age.max()
```

```python
standard_age.std()
```
