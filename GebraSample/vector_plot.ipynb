```python
import numpy as np
import matplotlib.pyplot as plt
```

# 2D

```python
u = np.array([2,4])
v = np.array([-2,2])
start = np.array([0,0])
R = u + v
```

```python
ax = plt.axes()
ax.set_xlim(-4,10)
ax.set_ylim(-4,10)
ax.grid(True)
ax.quiver(*start,*u,scale=1,angles = 'xy', scale_units = 'xy',color = 'red')
ax.quiver(*start,*v,scale=1,angles = 'xy', scale_units = 'xy',color = 'blue')
ax.quiver(*start,*R,scale=1,angles = 'xy', scale_units = 'xy',color = 'black')


plt.show()
```

# 3D

```python
u = np.array([2,4,4])
v = np.array([-2,2,3])
start = np.array([0,0,0])
R = u + v
```

```python
ax = plt.axes(projection = '3d')
ax.set_xlim(-4,10)
ax.set_ylim(-4,10)
ax.set_zlim(0,10)

ax.grid(True)
ax.quiver(*start,*u ,color = 'red')
ax.quiver(*start,*v ,color = 'blue')
ax.quiver(*start,*R ,color = 'black')

plt.show()
```

```python
A = np.array([[-1,2],[1.5,-3]])
```

```python
a1 = A.T[0]
a2 = A.T[1]
```

```python
a2
```

```python
np.linalg.det(A)
```

```python
u = np.array([2,2.5])
v = np.array([-1,4])
start = np.array([0,0])
```

```python
ut = np.matmul(A,u)
vt = np.matmul(A,v)
```

```python
ut
```

```python
vt
```

```python
ax = plt.axes()
ax.set_xlim(-2,10)
ax.set_ylim(-14,5)
ax.grid(True)
ax.quiver(*start,*u,scale=1,angles = 'xy', scale_units = 'xy',color = 'black')
ax.quiver(*start,*v,scale=1,angles = 'xy', scale_units = 'xy',color = 'black')

ax.quiver(*start,*a1,scale=1,angles = 'xy', scale_units = 'xy',color = 'green')
ax.quiver(*start,*a2,scale=1,angles = 'xy', scale_units = 'xy',color = 'green')

ax.quiver(*start,*ut,scale=1,angles = 'xy', scale_units = 'xy',color = 'red')
ax.quiver(*start,*vt,scale=1,angles = 'xy', scale_units = 'xy',color = 'blue')



plt.show()
```
